# Wine Quality Dataset Analysis

### Goal
The objective of this project is to analyze two wine datasets: one for red wine and one for white wine. We will focus on cleaning, tidying, and analyzing the data for insights.

### Dataset Overview
We are working with the **Wine Quality Dataset** from the UCI Machine Learning Repository. The dataset contains attributes related to the chemical properties of wine, along with a quality rating (the target variable). There are two datasets: one for red wine and one for white wine.

- **Features:**
  - Fixed acidity
  - Volatile acidity
  - Citric acid
  - Residual sugar
  - Chlorides
  - Free sulfur dioxide
  - Total sulfur dioxide
  - Density
  - pH
  - Sulphates
  - Alcohol
  - Quality (target variable)

The quality score is a rating between 0 and 10.


```python
import pandas as pd

# Paths to your local CSV files
red_wine_path = r"C:\Users\since\Downloads\wine+quality\winequality-red.csv"
white_wine_path = r"C:\Users\since\Downloads\wine+quality\winequality-white.csv"

# Load the datasets
red_wine_df = pd.read_csv(red_wine_path, delimiter=";")
white_wine_df = pd.read_csv(white_wine_path, delimiter=";")

# Display the first few rows of each dataset
print("Red Wine Data:")
print(red_wine_df.head())

print("\nWhite Wine Data:")
print(white_wine_df.head())
```

### DATA CLEANING:

```python
# Checking for missing values in both datasets
print("Missing values in Red Wine Dataset:")
print(red_wine_df.isnull().sum())

print("\nMissing values in White Wine Dataset:")
print(white_wine_df.isnull().sum())

# Checking the data types to ensure everything is correct
print("\nData types in Red Wine Dataset:")
print(red_wine_df.dtypes)

print("\nData types in White Wine Dataset:")
print(white_wine_df.dtypes)
```

## Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is the process of analyzing and summarizing datasets to gain insights, identify patterns, and detect anomalies or outliers. EDA is typically the first step in any data analysis or machine learning project. It helps to:

- Understand the structure and distribution of the data.
- Detect missing values or data quality issues.
- Identify relationships between variables.
- Highlight important trends, patterns, and anomalies.
- Guide further data analysis or model development.

Common EDA techniques include:
- **Summary Statistics**: Mean, median, standard deviation, etc., which provide a high-level overview of the dataset.
- **Visualizations**: Histograms, scatter plots, box plots, and bar charts to visualize distributions and relationships between variables.
- **Correlations**: Analyzing how different variables relate to each other, often using a correlation matrix.

In this project, we use EDA to explore the Wine Quality dataset, focusing on understanding the distribution of wine quality and how different chemical properties (e.g., alcohol, pH) relate to the quality ratings.


```python
import matplotlib.pyplot as plt

# Summary statistics for red wine
print("Summary Statistics for Red Wine:")
print(red_wine_df.describe())

# Summary statistics for white wine
print("\nSummary Statistics for White Wine:")
print(white_wine_df.describe())

# Plotting the distribution of wine quality for red wine
plt.figure(figsize=(6, 4))
red_wine_df['quality'].value_counts().sort_index().plot(kind='bar', color='red')
plt.title('Distribution of Red Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()

# Plotting the distribution of wine quality for white wine
plt.figure(figsize=(6, 4))
white_wine_df['quality'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Distribution of White Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

## Visualizing Relationships

Visualizing relationships between variables is a key part of Exploratory Data Analysis (EDA). It helps us understand how different features interact with each other and how they may influence the target variable.

- **Scatter Plots**: Used to identify correlations between two continuous variables. For example, a scatter plot of `alcohol` vs `quality` helps us see if higher alcohol content is associated with higher quality wine.
- **Box Plots**: Help visualize the distribution of a feature and detect outliers. By plotting `pH` against `quality`, we can observe the variation in pH levels across different quality ratings.
- **Bar Charts**: Useful for visualizing the distribution of categorical variables or counts, such as the number of wines that fall into each quality category.

These visualizations give us a clearer understanding of patterns in the data, guiding further analysis or modeling.


```python
# Scatter plot for Alcohol vs Quality in red wine
plt.figure(figsize=(8, 6))
plt.scatter(red_wine_df['alcohol'], red_wine_df['quality'], color='red', alpha=0.5)
plt.title('Red Wine: Alcohol vs Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()

# Scatter plot for Alcohol vs Quality in white wine
plt.figure(figsize=(8, 6))
plt.scatter(white_wine_df['alcohol'], white_wine_df['quality'], color='green', alpha=0.5)
plt.title('White Wine: Alcohol vs Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()

# Box plot for pH vs Quality in red wine
plt.figure(figsize=(8, 6))
red_wine_df.boxplot(column='pH', by='quality', grid=False, patch_artist=True)
plt.title('Red Wine: pH vs Quality')
plt.suptitle('')
plt.xlabel('Quality')
plt.ylabel('pH')
plt.show()

# Box plot for pH vs Quality in white wine
plt.figure(figsize=(8, 6))
white_wine_df.boxplot(column='pH', by='quality', grid=False, patch_artist=True)
plt.title('White Wine: pH vs Quality')
plt.suptitle('')
plt.xlabel('Quality')
plt.ylabel('pH')
plt.show()
```

### Conclusion
From our analysis of the red and white wine datasets, we have observed the following:
- The distribution of wine quality for both red and white wines is centered around 5-7, indicating that most wines in the dataset are of average quality.
- There seems to be a positive correlation between alcohol content and wine quality, meaning wines with higher alcohol content tend to receive higher quality ratings.
- The pH levels are fairly consistent across different quality ratings for both red and white wines.

Further analysis could explore additional relationships between other features (like acidity or residual sugar) and wine quality, and we could potentially build a machine learning model to predict wine quality based on these features.


# Titanic Dataset Analysis

### Goal
This analysis aims to explore the Titanic dataset, focusing on understanding the factors associated with passenger survival. We analyze demographic features like age, gender, and class, and their relationship with survival.

### Dataset Overview
The Titanic dataset includes the following features:
- PassengerId
- Pclass (Ticket class)
- Name
- Sex
- Age
- SibSp (Number of siblings/spouses aboard)
- Parch (Number of parents/children aboard)
- Ticket
- Fare
- Cabin
- Embarked (Port of embarkation)
- Survived (Target variable, 0 = No, 1 = Yes)


```python
import seaborn as sns
import pandas as pd

# Load the Titanic dataset from Seaborn
titanic_df = sns.load_dataset('titanic')

# Display the first few rows
print("Titanic Data:")
print(titanic_df.head())
```

```python
# Check for missing values
print("Missing values in Titanic Dataset:")
print(titanic_df.isnull().sum())

# Data type validation
print("\nData types in Titanic Dataset:")
print(titanic_df.dtypes)
```

```python
import matplotlib.pyplot as plt

# Bar plot for survival count
plt.figure(figsize=(6, 4))
titanic_df['survived'].value_counts().plot(kind='bar', color=['salmon', 'skyblue'])
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Scatter plot for Age vs Fare
plt.figure(figsize=(8, 6))
plt.scatter(titanic_df['age'], titanic_df['fare'], alpha=0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

### Conclusion
From the Titanic dataset analysis:
- The majority of passengers did not survive.
- Age and fare show a wide range, with younger passengers in both lower and higher fare categories.
Further analysis could include exploring survival rates by passenger class and gender.


# Iris Dataset Analysis

### Goal
This analysis aims to explore the relationships between sepal and petal measurements and the species of iris flowers. The goal is to identify patterns and characteristics that differentiate the species.

### Dataset Overview
The Iris dataset contains 150 samples with the following features:
- Sepal length
- Sepal width
- Petal length
- Petal width
- Species (Setosa, Versicolor, Virginica)


```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the species column
iris_df['species'] = iris.target

# Display the first few rows
print("Iris Data:")
print(iris_df.head())
```

```python
# Check for missing values
print("Missing values in Iris Dataset:")
print(iris_df.isnull().sum())

# Data type validation
print("\nData types in Iris Dataset:")
print(iris_df.dtypes)
```

```python
import matplotlib.pyplot as plt

# Distribution of Species
plt.figure(figsize=(6, 4))
iris_df['species'].value_counts().plot(kind='bar', color='green')
plt.title('Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

# Scatter plot for Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_df['species'], cmap='viridis', alpha=0.7)
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Species')
plt.show()
```

### Conclusion
From the Iris dataset analysis:
- Each species is evenly distributed with 50 samples.
- The scatter plot of sepal length vs. width shows some separation between species, indicating that these features might be useful in classification.
Further analysis could include exploring petal dimensions to better differentiate the species.
